```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Excel file (update path if needed)
df = pd.read_excel('tech_usage_survey_data.xlsx')

df.head()
```

```python
edu_hours = df.groupby('Education')['HoursPerDay'].mean().sort_values()
edu_hours.plot(kind='bar', color='skyblue')
plt.title('Avg Tech Hours by Education')
plt.ylabel('Hours Per Day')
plt.xlabel('Education Level')
plt.show()
```

```python
sns.barplot(data=df, x='Gender', y='SatisfactionScore',errorbar=None)
plt.title('Satisfaction by Gender')
plt.ylabel('Avg Satisfaction')
plt.show()
```

```python
crosstab = pd.crosstab(df['DeviceUsed'], df['WorkFromHome'])
crosstab.plot(kind='bar', stacked=True)
plt.title('Device Usage by Work-from-Home Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
sns.boxplot(data=df, x='UsesAIApps', y='TechComfortLevel')
plt.title('Tech Comfort Level: AI Users vs Non-users')
plt.show()
```
